## OpenCV 繪圖

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
# 畫直線

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')
# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
# 畫直線 + 矩形

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')

# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 矩形
#cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), -1)
cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), 5)

# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
# 畫直線 + 矩形 + 圓形

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')

# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 矩形
#cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), -1)
cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), 5)

# 圓形
#cv2.circle(im, (400, 100), 50, (0, 0, 255), -1)
cv2.circle(im, (400, 100), 50, (0, 0, 255), 6)


# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
# 畫直線 + 矩形 + 圓形 + 橢圓

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')

# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 矩形
#cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), -1)
cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), 5)

# 圓形
#cv2.circle(im, (400, 100), 50, (0, 0, 255), -1)
cv2.circle(im, (400, 100), 50, (0, 0, 255), 6)

# 橢圓形
cv2.ellipse(im, (100, 300), (60, 40), 135, 0, 360, (255, 255, 0), 10)


# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
# 畫直線 + 矩形 + 圓形 + 橢圓 + 多邊形

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')

# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 矩形
#cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), -1)
cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), 5)

# 圓形
#cv2.circle(im, (400, 100), 50, (0, 0, 255), -1)
cv2.circle(im, (400, 100), 50, (0, 0, 255), 6)

# 橢圓形
cv2.ellipse(im, (100, 300), (60, 40), 135, 0, 360, (255, 255, 0), 10)

# 多邊形
points = np.array([[200, 220], [220, 350], [280, 320], [300, 250]])
cv2.polylines(im, [points], True, (255, 0, 255), 10)

# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
# 畫直線 + 矩形 + 圓形 + 橢圓 + 多邊形 + 文字

# 影像畫布 (黑)
im = np.zeros([400, 500, 3], dtype='uint8')

# 直線
cv2.line(im, (50, 50), (150, 150), (255, 0, 0), 2, cv2.LINE_AA, 0)

# 矩形
#cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), -1)
cv2.rectangle(im, (200, 50), (300, 150), (0, 255, 0), 5)

# 圓形
#cv2.circle(im, (400, 100), 50, (0, 0, 255), -1)
cv2.circle(im, (400, 100), 50, (0, 0, 255), 6)

# 橢圓形
cv2.ellipse(im, (100, 300), (60, 40), 135, 0, 360, (255, 255, 0), 10)

# 多邊形
points = np.array([[200, 220], [220, 350], [280, 320], [300, 250]])
cv2.polylines(im, [points], True, (255, 0, 255), 10)

# 文字
text = 'Hello world'
font1 = cv2.FONT_HERSHEY_PLAIN
font2 = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(im, text, (300, 250), font1, 1.0, (255, 255, 255))
cv2.putText(im, text, (300, 280), font1, 2.0, (255, 255, 255))
cv2.putText(im, text, (300, 320), font2, 1.0, (255, 255, 255))
cv2.putText(im, text, (300, 360), font2, 1.0, (255, 255, 255), 3)


# 顯示 
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```
