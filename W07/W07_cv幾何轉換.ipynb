## 影像處理的數學

```python
# 影像縮放
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
```

```python
im = cv2.imread('img/Lenna.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
nr, nc = im.shape[:2]
nr2 = nr*2
nc2 = nc*2

im2 = cv2.resize(im, (nr2, nc2), interpolation=cv2.INTER_LINEAR)
plt.imshow(im2)
plt.show()
```

```python
# 影像旋轉

%matplotlib inline
import matplotlib.pyplot as plt
import cv2

im = cv2.imread('img/Lenna.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
nr2, nc2 = im.shape[:2]
rot_mat = cv2.getRotationMatrix2D((nc/2, nr/2), 30, 1)
im2 = cv2.warpAffine(im, rot_mat, (nc2, nr2) )

plt.imshow(im2)
plt.show()
```

```python
# 影像翻轉

%matplotlib inline
import matplotlib.pyplot as plt
import cv2

im = cv2.imread('img/Lenna.jpg')
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im)
plt.show()
```

```python
nr2, nc2 = im.shape[:2]
im2 = cv2.flip(im, 0)

plt.imshow(im2)
plt.show()
```

```python
nr2, nc2 = im.shape[:2]
im2 = cv2.flip(im, 1)

plt.imshow(im2)
plt.show()
```
